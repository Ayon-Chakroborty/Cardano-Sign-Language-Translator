```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from torch.autograd import Variable
```

```python
onnx_model_path = "/home/ayon_chakroborty/Desktop/modelResNet6.onnx"
sample_image = "/home/ayon_chakroborty/Desktop/Cardano DNN/test_B.jpg"
img = cv2.imread(sample_image)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img); plt.title("Sign Language Letter: B")
```

```python
net = cv2.dnn.readNetFromONNX(onnx_model_path) 
print(net)
blob = cv2.dnn.blobFromImage(img, scalefactor=1, size=(225,225), mean=(0.229, 0.224, 0.225), swapRB = False, crop=False)
net.setInput(blob)
print(blob.shape)
```
